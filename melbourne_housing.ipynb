```python
#z-score
from scipy import stats
z=np.abs(stats.zscore(boston_df))
```

```python
print(z)
```

```python
threshold=3
np.where(z>threshold)
#WE ARE FINDING VALUES GREATERTHAN 3 ,1ST ARRAY GIVE ROWS VALUS AND 12ND ARRAY GIVE COLUMN VALUE
```

```python
#for checking the value from above table we can just check it by giving row and column value
print(z.iloc[55][1])
```

```python
#Iqr
Q1=boston_df.quantile(0.25)
Q3=boston_df.quantile(0.75)
IQR=Q3-Q1
```

```python
print(IQR)
```

```python
np.where((boston_df<(Q1-1.5*IQR))|(boston_df>(Q3+1.5*IQR)))
```

```python
print(z.loc[18][11])
```

```python
#correcting outliers
#zscore
```

```python
boston_df1=boston_df[(z<3).all(axis=1)]
print(boston_df.shape)
print(boston_df1.shape)
```

```python
boston_df2=boston_df.copy()
```

```python
#replace outliers with median value
for i,j in zip(np.where(z>threshold)[0],np.where(z>threshold)[1]):
    boston_df2.iloc[i,j]=boston_df.iloc[:,j].median()
```

```python
#IQR
#replace every outliers on the lower side by the lower whisker
for i,j in zip(np.where(boston_df<Q1-1.5*IQR)[0],np.where(boston_df<Q1-1.5*IQR)[1]):
    whisker=Q1-1.5*IQR-
    boston_df.iloc[i,j]=whisker[j]
```

```python
#replace every outliers on the upper side by the upper whisker
for i,j in zip(np.where(boston_df>Q3+1.5*IQR)[0],np.where(boston_df>Q3+1.5*IQR)[1]):
    whisker=Q3+1.5*IQR
    boston_df.iloc[i,j]=whisker[j]
```
